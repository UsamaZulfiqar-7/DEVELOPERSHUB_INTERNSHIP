### Problem Statement
The Iris dataset is one of the most well-known datasets in machine learning and statistics. It contains measurements of 150 iris flowers from three different species: **Setosa**, **Versicolor**, and **Virginica**. The goal of this task is to explore the dataset structure and create meaningful visualizations to understand relationships between features.

### Dataset Description
- **Source**: Seaborn library (originally from UCI ML Repository)
- **Rows**: 150 samples | **Columns**: 5
- **Features**: sepal_length, sepal_width, petal_length, petal_width (all in cm)
- **Target**: species (Setosa, Versicolor, Virginica)


## 1. Import Libraries

```python
# Import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set plot style
sns.set_theme(style="whitegrid", palette="Set2")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

print("Libraries imported successfully!")
```

## 2. Load Dataset

```python
# Load the Iris dataset from CSV
df = pd.read_csv('iris.csv')

print("Dataset loaded successfully!")
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
```

## 3. Dataset Structure & Understanding

```python
# Display column names
print("Column Names:")
print(df.columns.tolist())
```

```python
# Display first 10 rows
print("First 10 Rows:")
df.head(10)
```

```python
# Dataset shape
print(f"Shape: {df.shape}")
print(f"Total Records: {df.shape[0]}")
print(f"Total Features: {df.shape[1]}")
```

```python
# Data types and non-null counts
print("Dataset Info:")
df.info()
```

```python
# Statistical summary
print("Statistical Summary:")
df.describe().round(2)
```

```python
# Check missing values
print("Missing Values per Column:")
print(df.isnull().sum())
print(f"\nTotal Missing Values: {df.isnull().sum().sum()}")
```

```python
# Class distribution
print("Species Distribution:")
print(df['species'].value_counts())
print()
print("Percentage Distribution:")
print(df['species'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%')
```

## 4. Exploratory Data Analysis (EDA)

```python
# Mean feature values by species
print("Mean Feature Values by Species:")
df.groupby('species').mean().round(2)
```

## 5. Visualizations
### 5.1 Scatter Plots – Relationship Between Variables

```python
# Scatter plot: Sepal Length vs Sepal Width
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Sepal
sns.scatterplot(data=df, x='sepal_length', y='sepal_width',
                hue='species', palette='Set1', s=80, ax=axes[0])
axes[0].set_title('Sepal Length vs Sepal Width', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Sepal Width (cm)')

# Plot 2: Petal
sns.scatterplot(data=df, x='petal_length', y='petal_width',
                hue='species', palette='Set1', s=80, ax=axes[1])
axes[1].set_title('Petal Length vs Petal Width', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Petal Length (cm)')
axes[1].set_ylabel('Petal Width (cm)')

plt.suptitle('Scatter Plots: Feature Relationships by Species', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('scatter_plots.png', dpi=150, bbox_inches='tight')
plt.show()
print("Observation: Petal dimensions show much clearer species separation than sepal dimensions.")
```

### 5.2 Histograms – Data Distribution

```python
# Histograms for all features
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, feature in enumerate(features):
    for species in df['species'].unique():
        subset = df[df['species'] == species][feature]
        axes[i].hist(subset, bins=15, alpha=0.6, label=species, edgecolor='white')
    axes[i].set_title(f'Distribution of {feature.replace("_", " ").title()}', fontsize=13, fontweight='bold')
    axes[i].set_xlabel(f'{feature.replace("_"," ").title()} (cm)')
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.suptitle('Histograms: Feature Distributions by Species', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('histograms.png', dpi=150, bbox_inches='tight')
plt.show()
print("Observation: Petal length and width show clear bimodal distributions separating Setosa from others.")
```

### 5.3 Box Plots – Outlier Detection & Spread

```python
# Box plots for all features by species
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()
colors = ['#E74C3C', '#2ECC71', '#3498DB']

for i, feature in enumerate(features):
    sns.boxplot(data=df, x='species', y=feature, palette=colors, ax=axes[i])
    axes[i].set_title(f'Box Plot: {feature.replace("_"," ").title()} by Species',
                      fontsize=13, fontweight='bold')
    axes[i].set_xlabel('Species')
    axes[i].set_ylabel(f'{feature.replace("_"," ").title()} (cm)')

plt.suptitle('Box Plots: Feature Spread & Outliers by Species', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('box_plots.png', dpi=150, bbox_inches='tight')
plt.show()
print("Observation: Setosa has very distinct and compact feature ranges. Versicolor and Virginica overlap slightly.")
```

### 5.4 Pair Plot – All Feature Combinations

```python
# Pair plot to see all feature relationships
g = sns.pairplot(df, hue='species', palette='Set1', 
                 diag_kind='kde', plot_kws={'alpha': 0.7, 's': 50})
g.fig.suptitle('Pair Plot: All Feature Combinations', y=1.02, fontsize=16, fontweight='bold')
plt.savefig('pair_plot.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 5.5 Correlation Heatmap

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
corr = df[features].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, square=True, linewidths=0.5,
            cbar_kws={'shrink': 0.8})
plt.title('Correlation Heatmap of Iris Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("Observation: Petal length and petal width are highly correlated (r=0.96).")
```

## 6. Conclusion

### Key Insights from the Iris Dataset Exploration:

1. **Dataset Quality**: The Iris dataset is clean with no missing values and perfectly balanced classes (50 samples each).

2. **Best Discriminating Features**: Petal length and petal width are the most powerful features for distinguishing species, with very clear cluster separation visible in scatter plots.

3. **Setosa is Highly Distinct**: Iris Setosa has noticeably smaller petals and forms a completely separate cluster from Versicolor and Virginica.

4. **Versicolor vs Virginica**: These two species show some overlap, especially in sepal measurements, making them harder to distinguish.

5. **Feature Correlations**: Petal length and petal width are strongly correlated (r = 0.96), suggesting that only one of them may be needed in a predictive model.

6. **Outliers**: A few outliers exist in sepal width for Setosa (as seen in box plots), but they are minimal.
